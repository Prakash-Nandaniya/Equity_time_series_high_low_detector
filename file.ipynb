```python
import pandas as pd
import numpy as np

# File path
input_file = "modified_train.csv"

# Load the CSV into a Pandas DataFrame
df = pd.read_csv(input_file)
df['class_label'] = df['class_label'].fillna('Nan').replace('', 'Nan')
# Check for ticker_id distribution
print("=== TICKER_ID CHECK ===")
print(f"Unique ticker_ids: {df['ticker_id'].nunique()}")
print(f"Ticker_ids: {sorted(df['ticker_id'].unique())}")
print(f"Count per ticker_id:")
print(df['ticker_id'].value_counts().sort_index())

# Check if ticker_id is string or needs conversion
if df['ticker_id'].dtype != 'object':
    print("Converting ticker_id to string...")
    df['ticker_id'] = df['ticker_id'].astype(str).str.zfill(3)

print(f"After conversion - Unique ticker_ids: {df['ticker_id'].nunique()}")
print(f"After conversion - Ticker_ids: {sorted(df['ticker_id'].unique())}")
print(f"After conversion - Count per ticker_id:")
print(df['ticker_id'].value_counts().sort_index())

# columns to average
average_cols = ['momentum', 'ratio', 'sm_momentum', 'sm_ratio', 'cross_threshold_from_above', 'cross_threshold_from_below', 'trending_down_and_above', 'trending_up_and_below', 'troughs', 'zone', 'peaks']

# Check if all columns exist
missing_cols = [col for col in average_cols if col not in df.columns]
if missing_cols:
    print(f"WARNING: Missing columns: {missing_cols}")
    average_cols = [col for col in average_cols if col in df.columns]

# Centralize columns by subtracting mean
df[average_cols] = df[average_cols] - df[average_cols].mean()
print(f"Centered {len(average_cols)} columns: {average_cols}")

# Columns to exclude from logarithmic transformation 
log_cols = ['momentum', 'ratio', 'sm_momentum', 'sm_ratio']

# Check if log columns exist
missing_log_cols = [col for col in log_cols if col not in df.columns]
if missing_log_cols:
    print(f"WARNING: Missing log columns: {missing_log_cols}")
    log_cols = [col for col in log_cols if col in df.columns]

# Apply logarithmic transformation to specified columns
for col in log_cols:
    df[col] = np.sign(df[col]) * np.log1p(np.abs(df[col]))

print(f"Applied log transformation to {len(log_cols)} columns: {log_cols}")

# Final ticker_id check
print("\n=== FINAL TICKER_ID CHECK ===")
print(f"Unique ticker_ids: {df['ticker_id'].nunique()}")
print(f"Count per ticker_id:")
print(df['ticker_id'].value_counts().sort_index())

print(f"\nUpdated df with mean-subtracted values and logarithmic transformation applied to {log_cols}")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Assume df_001 is already defined in the environment
df_001 = df_001.copy()  # Create copy to avoid SettingWithCopyWarning

# Handle NaN in class_label by mapping to 'Nan'
df_001.loc[:, 'class_label'] = df_001['class_label'].fillna('Nan')

# Feature columns: only the specified ones for NN
feature_cols = ['sm_momentum', 'sm_ratio', 'momentum', 'ratio', 'peaks', 'troughs']

# Handle any NaN in features (fill with 0 using .loc)
df_001.loc[:, feature_cols] = df_001[feature_cols].fillna(0)

# Create second letter labels: Nan, H (for HH, LH), L (for LL, HL)
def get_second_letter(label):
    if label == 'Nan':
        return 'Nan'
    elif label in ['HH', 'LH']:
        return 'H'
    elif label in ['LL', 'HL']:
        return 'L'
    return 'Nan'  # Fallback

df_001.loc[:, 'second_letter'] = df_001['class_label'].apply(get_second_letter)

# Create first letter labels for non-Nan: H or L
def get_first_letter(label):
    if label == 'Nan':
        return None
    elif label in ['HH', 'HL']:
        return 'H'
    elif label in ['LL', 'LH']:
        return 'L'
    return None  # Fallback

df_001.loc[:, 'first_letter'] = df_001['class_label'].apply(get_first_letter)

# Label encoders
le_second = LabelEncoder()
le_second.fit(df_001['second_letter'].unique())
print("Second letter mapping:", dict(zip(le_second.classes_, range(len(le_second.classes_)))))

le_first = LabelEncoder()
non_nan_first = df_001[df_001['first_letter'].notna()]['first_letter'].unique()
le_first.fit(non_nan_first)
print("First letter mapping:", dict(zip(le_first.classes_, range(len(le_first.classes_)))))

# Split data: Train on all data, then select 50% for test
X = df_001[feature_cols].values.astype(np.float32)
y_second_encoded = le_second.transform(df_001['second_letter'])
y_second = to_categorical(y_second_encoded, num_classes=len(le_second.classes_))

n_samples = len(df_001)
test_idx = np.random.choice(n_samples, size=int(n_samples * 0.5), replace=False)
train_idx = np.setdiff1d(np.arange(n_samples), test_idx)

X_train = X[train_idx]
X_test = X[test_idx]
y_train = y_second[train_idx]
y_test = y_second[test_idx]

# === NN for Second Character ===
# Train NN on all data
model_second = Sequential()
model_second.add(Dense(64, input_dim=len(feature_cols), activation='relu'))
model_second.add(Dense(32, activation='relu'))
model_second.add(Dense(len(le_second.classes_), activation='softmax'))

model_second.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history_second = model_second.fit(X_train, y_train, epochs=30, batch_size=16, verbose=1, validation_split=0.2)

# Predict second letter with NN
y_second_pred_prob_nn = model_second.predict(X_test)
y_second_pred_nn = np.argmax(y_second_pred_prob_nn, axis=1)

# === Voting for Second Character ===
def vote_for_second(row):
    votes = {'H': 0, 'L': 0}
    
    # sm_momentum
    if row['sm_momentum'] < -0.7:
        votes['L'] += 1
    elif row['sm_momentum'] > 0.2:
        votes['H'] += 1
    
    # sm_ratio
    if row['sm_ratio'] < -0.07:
        votes['L'] += 1
    elif row['sm_ratio'] > 0.07:
        votes['H'] += 1
    
    # momentum
    if row['momentum'] < -0.07:
        votes['L'] += 1
    elif row['momentum'] > 0.07:
        votes['H'] += 1
    
    # ratio
    if row['ratio'] < -0.07:
        votes['L'] += 1
    elif row['ratio'] < -0.1:
        votes['L'] += 2
    elif row['ratio'] > 0.07:
        votes['H'] += 1
    elif row['ratio'] > 0.1:
        votes['H'] += 2
        
    # peaks
    if row['peaks'] > 0.7:
        votes['H'] += 1
    
    # troughs
    if row['troughs'] > 0.07:
        votes['L'] += 1
    elif row['troughs'] < -0.07:
        votes['H'] += 1
        
    if row['trending_up_and_below'] < -0.13:
        votes['L'] += 1
        votes['H'] += 1
    
    # Voting logic: need at least 4 votes for a class
    if votes['H'] > votes['L'] and votes['H'] >= 6:
        return 'H'
    elif votes['L'] > votes['H'] and votes['L'] >= 3:
        return 'L'
    else:
        return 'Nan'

# Apply voting to test set
test_df = df_001.iloc[test_idx].copy()
test_df.loc[:, 'predicted_second_vote'] = test_df.apply(vote_for_second, axis=1)

# Combine NN and Voting (use NN if confidence > 0.99, else voting)
test_df.loc[:, 'predicted_second'] = le_second.inverse_transform(y_second_pred_nn)
confidence_threshold = 0.99
y_second_confidence = np.max(y_second_pred_prob_nn, axis=1)
test_df.loc[y_second_confidence < confidence_threshold, 'predicted_second'] = test_df.loc[y_second_confidence < confidence_threshold, 'predicted_second_vote']

# === NN for First Character (if not Nan) ===
non_nan_train = df_001[df_001['class_label'] != 'Nan']
y_first_encoded_train = le_first.transform(non_nan_train['first_letter'])
X_train_nn = non_nan_train[feature_cols].values.astype(np.float32)
y_train_nn = to_categorical(y_first_encoded_train, num_classes=2)

model_first = Sequential()
model_first.add(Dense(64, input_dim=len(feature_cols), activation='relu'))
model_first.add(Dense(32, activation='relu'))
model_first.add(Dense(2, activation='softmax'))

model_first.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history_first = model_first.fit(X_train_nn, y_train_nn, epochs=30, batch_size=16, verbose=1, validation_split=0.2)

# Predict first letter for test set where predicted_second != 'Nan'
non_nan_test = test_df[test_df['predicted_second'] != 'Nan']
if not non_nan_test.empty:
    X_test_nn = non_nan_test[feature_cols].values.astype(np.float32)
    y_first_pred_prob = model_first.predict(X_test_nn)
    y_first_pred = np.argmax(y_first_pred_prob, axis=1)
    first_letters = le_first.inverse_transform(y_first_pred)
    test_df.loc[test_df['predicted_second'] != 'Nan', 'predicted_first'] = first_letters
else:
    test_df['predicted_first'] = None

# Final prediction: combine first and second if not Nan, else Nan
def get_final_pred(row):
    if row['predicted_second'] == 'Nan':
        return 'Nan'
    else:
        return row['predicted_first'] + row['predicted_second']

test_df.loc[:, 'predicted_label'] = test_df.apply(get_final_pred, axis=1)

# === Evaluation ===
y_true = test_df['class_label']
le_full = LabelEncoder()
le_full.fit(['Nan', 'LL', 'LH', 'HH', 'HL'])
y_true_encoded = le_full.transform(y_true)
y_pred_encoded = le_full.transform(test_df['predicted_label'])

cm = confusion_matrix(y_true_encoded, y_pred_encoded, labels=range(5))

fig, ax = plt.subplots(1, 1, figsize=(5, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=le_full.classes_, yticklabels=le_full.classes_,
            ax=ax, cbar=False)
ax.set_title(f'Confusion Matrix - Ticker 001\n(Accuracy: {np.mean(y_pred_encoded == y_true_encoded):.3f})')
ax.set_xlabel('Predicted Label')
ax.set_ylabel('True Label')
plt.tight_layout()
plt.show()

print("\n=== PREDICTION STATISTICS FOR TICKER 001 ===")
print(f"Accuracy: {np.mean(y_pred_encoded == y_true_encoded):.4f}")
print(f"Predicted counts: {dict(zip(le_full.classes_, np.bincount(y_pred_encoded, minlength=5)))}")
print(f"Actual counts: {dict(zip(le_full.classes_, np.bincount(y_true_encoded, minlength=5)))}")
```
